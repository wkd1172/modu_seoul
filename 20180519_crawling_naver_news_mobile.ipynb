```python
def url_params(url):
    data = {}
    url_main,params = url.split("?",1)
    params_2 = params.replace("&",",").replace(":","=").split(",")
    for index in params_2:
        key_a,value_a = index.split("=",1)
        if value_a == "null":
            value_a = None
        data[key_a] = value_a
    return data
```

```python
def NNC(search):
    title_2 = []
    date_2 = []
    url = "https://m.search.naver.com/search.naver"
    headers = {
        "Referer" : "https://m.naver.com/"
    }
    for page in range(1,10):
        params = {
            'de': '',
            'ds': '',
            'field': '0',
            'mynews': '0',
            'news_office_checked': '',
            'nso': 'so=r',
            'office_section_code': '0',
            'office_type': '0',
            'p': 'all',
            'pd': '0',
            'photo': '0',
            'query': search,
            'sm': 'mtb_pge',
            'sort': '1',
            'start': (page-1)*15 +1,
            'where': 'm_news'
            }
        html = requests.get(url,params = params,headers = headers).text
        soup = BeautifulSoup(html,"html.parser")
        for tag in soup.select(".list_news .bx a"):
            #title = tag.find("div",class_ = "news_tit").text
            #date_1 = tag.findAll("div",class_ = "news_dsc")
            #date_2 = date_1.findAll("div",class_= "dsc_wrap")
            #date_3 = date_2.findAll("div",class_="elss dsc_sub")
            #date = date_3.findAll("span").text
            #date = tag.find("span", class_ = "sub_txt sub_time").text
            #key_a,value_a = date,title
            #data[key_a] = value_a
            for tag_2 in tag.select(".news_tit"):
                if tag_2 == "null":
                    tag_2 = None
                title_2.append(tag_2.text)
                
            for tag_2 in tag.select("div span"):
                if tag_2 == "null":
                    tag_2 = None
                date_2.append(tag_2.text)
        data_frame = pd.DataFrame({
        "title" : title_2,
        "date" : date_2}
        )

    return data_frame
            #news_title.append(title)
            

```
