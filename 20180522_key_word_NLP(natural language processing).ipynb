```python
kkma = Kkma()
data = []

for tag in range(0,2000):
    data.append(kkma.nouns(data_base["title"][tag]))
data
```

```python
all_data =[]
for tag in data:
    #data_1 = []
    for count in tag:
        all_data.append(count)
    
all_data
```
