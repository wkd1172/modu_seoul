```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
https://bigdata.seoul.go.kr/share/selectPageListShare.do?r_id=P410&cnrs_seq=&sch_task_cl_code=&sch_text=&currentPage=2
```

```python
def crawling_machine(mid_class):
    data = []
    if mid_class == "resident":
        url = "https://bigdata.seoul.go.kr/share/selectPageListShare.do?r_id=P410"
        for page in range(1,19):
            params = {
                "currentPage" : page
            }
            html = requests.get(url,params = params).text
            soup = BeautifulSoup(html,"html.parser")
            for tag in soup.select(".board_list tbody tr td"):
                data.append(tag.text)
    
        no = []
        Mid_class = []
        title = []
        data_type = []
        date = []

        for set in range(1,179):
            no.append(data[7*(set-1)])
            Mid_class.append(data[7*(set-1)+1])
            title.append(data[7*(set-1)+2])
            dt = [data[7*(set-1)+3:7*(set-1)+6]]
            data_type.append(dt)
            date.append(data[7*(set-1)+6])

        data_base = pd.DataFrame({
            "no" : no,
            "Mid_class" : Mid_class,
            "title" : title,
            "data_type" : data_type,
            "date" : date,}
            ,columns = {"Mid_class","title","data_type","date"},
            index = no)    
        return data_base
    
    elif mid_class == "collaboration":
        data = []
        url = "https://bigdata.seoul.go.kr/noti/selectPageListNoti.do?r_id=P420"
        for page in range(1,4):
            params = {
                "currentPage" : page
            }
            html = requests.get(url,params = params).text
            soup = BeautifulSoup(html,"html.parser")
            for tag in soup.select(".board_list tbody tr td"):
                data.append(tag.text)
        no = []
        title = []
        date = []
        views = []
        
        for set in range(1,27):
            no.append(data[4*(set-1)])
            title.append(data[4*(set-1)+1])
            date.append(data[4*(set-1)+2])
            views.append(data[4*(set-1)+3])
        data_base = pd.DataFrame({
            "no" : no,
            "title" : title,
            "date" : date,
            "views" : views
            },
            columns = {"title","date","views"},
            index = no)
        return data_base
    elif mid_class == "contest":
        data = []
        url = "https://bigdata.seoul.go.kr/noti/selectPageListNoti.do?r_id=P440"
        for page in range(1,5):
            params ={
                "currentPage" : page
            }
            html = requests.get(url,params= params).text
            soup = BeautifulSoup(html,"html.parser")
            for tag in soup.select(".board_list tbody tr td"):
                data.append(tag.text)
        no = []
        title = []
        date = []
        views = []
        #38개
        for set in range(1,39):
            no.append(data[4*(set-1)])
            title.append(data[4*(set-1)+1])
            date.append(data[4*(set-1)+2])
            views.append(data[4*(set-1)+3])
        data_base = pd.DataFrame({
            "no" : no,
            "views" : views,
            "title" : title,
            "date" : date
            },
            columns = {"views","title","date",},
            index = no)
        return data_base
        
```

```python
Data_contest =crawling_machine("contest")
```

```python
Data_collaboration =crawling_machine("collaboration")
```

```python
Data_resident= crawling_machine("resident")
```

```python
Data_contest.to_csv("seoul_pro_contest.csv")
Data_resident.to_csv("seoul_pro_resident.csv")
Data_collaboration.to_csv("seoul_pro_collaboration.csv")
```
