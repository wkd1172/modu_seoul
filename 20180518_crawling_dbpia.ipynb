```python
def crawling_dbpia(search):
    url = "http://m.dbpia.co.kr/SearchResult/Search"
    title = []
    etc = []
    for page in count(1):
        params = {
            'Collection': '0',
            'Multimedia': '0',
            'Page': page,
            'SearchAll': search,
            'SearchAuthor': '',
            'SearchBook': '',
            'SearchDissertation': '',
            'SearchPublisher': '',
            'dpTypeName': '',
            'fCategory': '',
            'fContentsType': '',
            'fDpTypeCode': '',
            'fPublication': '',
            'fPublishYear': '',
            'fPublisher': '',
            'fRegister': '',
            'firstCategory': '',
            'fristCategoryName': '',
            'nPagesize': '20',
            'nSearchField': '',
            'nSearchKeyword': '',
            'nSort': '1',
            'nSorttype': 'desc',
            'publisherYearName': '',
            'q': '%28%5B147003%C2%A7coldb%C2%A73%C2%A712%C2%A73%5D%29+AND+%5B%EB%8F%84%EC%8B%9C%EB%AC%B8%EC%A0%9C%C2%A7coldb%C2%A72%C2%A751%C2%A73%5D',
            'searchReset': '',
            'searchWord': '%EC%A0%84%EC%B2%B4%3D%5E%24%EB%8F%84%EC%8B%9C%EB%AC%B8%EC%A0%9C%5E*',
            'searchWordCondition': '',
            'secondCategory': '',
            'secondCategoryName': '',
            'specificParam': ''
        }
        refer = {
            "Referer" : "http://m.dbpia.co.kr/SearchResult"
        }
        html = requests.get(url,params = params, headers = refer).text
        soup = BeautifulSoup(html,"html.parser")
        for tag in soup.select("#searchListView .ui-btn-text a div"):
            title_2 = tag.findAll(class_ = "search_result_title")
            etc_2 = tag.findAll(class_ = "search_result_etc")
            print(title_2)
            priint(etc_2)
            print("_+______________----")
            if title_2 in title:
                return title,etc
            else: 
                title.append(title_2)
                etc.append(etc_2)
        return title,etc
    
            
            
        
```
